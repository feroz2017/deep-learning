```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Math, Latex
```

```python
import sympy as sp
```

```python
x = sp.symbols('x')
f_sym = x**3 - 6*x**2 + 9*x +1
```

```python
print(sp.latex(f_sym))
```

```python
display(Math(sp.latex(f_sym)))
```

```python
f_prime_sym = sp.diff(f_sym)
```

```python
display(Math(sp.latex(f_prime_sym)))
```

```python
f_prime_prime_sym = sp.diff(f_prime_sym)
display(Math(sp.latex(f_prime_prime_sym)))
```
